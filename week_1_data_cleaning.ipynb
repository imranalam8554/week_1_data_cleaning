```python
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
categorical_cols = df.select_dtypes(include=['object', 'string']).columns

for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df['initial_price'] = pd.to_numeric(df['initial_price'], errors='coerce')
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
```

```python
selected_columns = df[['title', 'final_price', 'category']]
selected_columns.head()
```

```python
filtered_products = df[df['rating'] > 4]

filtered_products.head()
```

```python
print("Duplicates Before:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicates After:", df.duplicated().sum())
```

```python
df['quantity'] = np.random.randint(1, 10, size=len(df))
```

```python
df['total_amount'] = df['quantity']
```

```python
df['total_amount'] = df['rating'] * df['quantity']
```

```python
df.to_csv("cleaned_dataset.csv", index=False)

print("Cleaned dataset saved successfully.")
```
